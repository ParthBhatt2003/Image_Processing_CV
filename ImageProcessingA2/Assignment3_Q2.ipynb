Importing Required Libraries

```python
import numpy as np
import math
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
```

CODE FOR IMAGE 1

```python
# Taking image as an array
img=cv.imread('cctv1.JPG')
# cv2_imshow(img)

# Converting into gray scale
gray=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cv2_imshow(gray)

# Increasing the Contrast
equ = cv.equalizeHist(gray)
# cv2_imshow(equ)
final1=equ

res = np.hstack((gray,final1)) #stacking images side-by-side
cv2_imshow(res)
```

CODE FOR IMAGE 2

```python
# Taking image as an array
img=cv.imread('cctv2.JPG')
# cv2_imshow(img)

# Converting into gray scale
gray=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cv2_imshow(gray)

# Increasing the Brightness
image_shifted = cv.add(gray, 30)
image_shifted = np.clip(image_shifted, 0, 255)
# cv2_imshow(image_shifted)
final2=image_shifted

res = np.hstack((gray,final2)) #stacking images side-by-side
cv2_imshow(res)

# Increasing Contrast Produce destorted Image
```

CODE FOR IMAGE 3

```python
# Taking image as an array
img=cv.imread('cctv3.JPG')
# cv2_imshow(img)

# Converting into gray scale
gray=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cv2_imshow(gray)

# Increasing the Contrast
equ = cv.equalizeHist(gray)
final3=equ

res = np.hstack((gray,final3)) #stacking images side-by-side
cv2_imshow(res)
```

CODE FOR IMAGE 4

```python
# Taking image as an array
img=cv.imread('cctv4.JPG')
# cv2_imshow(img)

# Converting into gray scale
gray=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cv2_imshow(gray)

# Increasing the Contrast and reducing the Brightness
equ = cv.equalizeHist(gray)
image_shifted = cv.add(equ, -30)
image_shifted = np.clip(image_shifted, 0, 255)

final4=image_shifted

res = np.hstack((equ,final4)) #stacking images side-by-side
cv2_imshow(res)
```
