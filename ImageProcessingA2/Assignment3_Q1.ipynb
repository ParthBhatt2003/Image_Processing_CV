CODE FOR IMAGE 1

```python
# Taking image as an array
img=cv.imread('gutters1.JPG')
cv2_imshow(img)
```

```python
# Dilating the image to get the gutters
dilated_img = cv.dilate(img, np.ones((7,7), np.uint8))
cv2_imshow(dilated_img)

# median=cv.medianBlur(dilated_img, 17)
# cv2_imshow(median)
# bilateral=cv.bilateralFilter(sharpened,15,30,20)
# cv2_imshow(bilateral)
```

```python
# Removing all the gutters from the image
diff_img = 255 - cv.absdiff(img, dilated_img)
cv2_imshow(diff_img)

dilated_gray=cv.cvtColor(dilated_img, cv.COLOR_BGR2GRAY)
diff_gray=cv.cvtColor(diff_img, cv.COLOR_BGR2GRAY)
```

```python
#Improving the quality

# gauss=cv.GaussianBlur(sharpened,(3,3),20)
# # Sharpening the image
# kernel = np.array([[0, -1, 0],
# [-1, 5, -1],
# [0, -1, 0]])
# sharpened = cv.filter2D((thresh), -1, kernel)
# cv2_imshow(sharpened)

# Thresholding in two parts

# print(img.shape[1])
cropped=diff_gray[0:777, 0:80]
# cv2_imshow(cropped)
threshold, thresh=cv.threshold(cropped, 215, 255 ,cv.THRESH_BINARY)
# cv2_imshow(thresh)
diff_gray[0:777, 0:80]=thresh

cropped=diff_gray[0:777, 81:585]
threshold, thresh=cv.threshold(cropped, 210, 255 ,cv.THRESH_BINARY)
diff_gray[0:777, 81:585]=thresh

final1=diff_gray
cv2_imshow(final1)
```

```python
# Taking image as an array
img=cv.imread('gutters2.JPG')
# cv2_imshow(img)
```

```python
# Dilating the image to get the gutters

gray=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cv2_imshow(gray)

# print(gray.shape[0])

dilated_img = cv.dilate(gray, np.ones((5,5), np.uint8))
# cv2_imshow(dilated_img)

threshold, thresh=cv.threshold(dilated_img, 90, 255, cv.THRESH_BINARY)
# cv2_imshow(thresh)

diff_1=255-cv.absdiff(dilated_img,thresh)
cv2_imshow(diff_1)
```

```python
# Removing all the gutters from the image
diff_img = 255 - cv.absdiff(gray, diff_1)

final2=diff_img
cv2_imshow(final2)


# cropped=gray[0:935, 510:620]
# # cv2_imshow(cropped)

# threshold, thresh=cv.threshold(cropped, 100, 255, cv.THRESH_BINARY)
# cv2_imshow(thresh)
# gray[0:935, 510:620]=thresh
# cv2_imshow(gray)
```
